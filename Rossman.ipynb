### Preprocesamiento duraciones:

```python
joined_train = pd.read_feather('train_checkpoint_1.fth')
joined_test = pd.read_feather('test_checkpoint_1.fth')
```

```python
#State holiday a bool
joined_train['StateHoliday_bool'] = joined_train.StateHoliday!='0'
joined_test['StateHoliday_bool'] = joined_test.StateHoliday!='0'
```

```python
class elapsed(object):
    def __init__(self, fld):
        self.fld = fld
        self.last = pd.to_datetime(np.nan)
        self.last_store = 0
        
    def get(self, row):
        if row.Store != self.last_store:
            self.last = pd.to_datetime(np.nan)
            self.last_store = row.Store
        if (row[self.fld]): self.last = row.Date
        return row.Date-self.last
```

```python
#Defino Closed como opuesto a open. La idea es registrar d칤as y duraciones anteriores y posteriores a la tienda cerrada.
joined_train['Closed'] = (~joined_train['Open'].astype('bool')).astype('int32')
joined_test['Closed'] = (~joined_test['Open'].astype('bool')).astype('int32')
```

```python
columns = ["Date", "Store", "Promo", "StateHoliday_bool", "SchoolHoliday", "Closed"]
```

```python
dur_train = joined_train[columns]
dur_test = joined_test[columns]
```

```python
def add_elapsed(df, fld, prefix):
    sh_el = elapsed(fld)
    df[prefix+fld] = df.apply(sh_el.get, axis=1)
```

### Predicci칩n de Sales y Customers usando Transformada de Fourier:

```python
from FourierRegressor import FourierRegressor
```

```python
def train_store_regressors (df, store, n=5):
    store_data = df[df["Store"] == store]
    sales_regressor = FourierRegressor()
    cust_regressor = FourierRegressor()
    t = store_data["Date"] - df["Date"].min()
    t = t.dt.days
    sales = store_data["Sales"].values
    customers = store_data["Customers"].values
    sales_regressor.fit(t, sales, n, low_freq_filter=0.05)
    cust_regressor.fit(t, customers, n, low_freq_filter=0.05)
    
    return sales_regressor, cust_regressor
```

```python
def predict_store_output (df, store, sales_regressor, cust_regressor):
    store_data = df[df["Store"] == store]
    t = store_data["Date"] - df["Date"].min()
    t = t.dt.days
    return sales_regressor.predict(t), cust_regressor.predict(t)
```

```python
store_1088 = joined_train[joined_train["Store"] == 1088]
sales_reg, cust_reg = train_store_regressors(joined_train, 1088)
pred_sales, pred_cust = predict_store_output(joined_train, 1088, sales_reg, cust_reg)
```

```python
def calculate_store_predictions(df, df_train, n = 5, print_progress_step = 100):
    df["SalesPred"] = np.zeros(len(df))
    df["CustPred"] = np.zeros(len(df))
    stores_done = 0
    for store in df["Store"].unique():
        sales_reg, cust_reg = train_store_regressors(df_train, store, n)
        pred_sales, pred_cust = predict_store_output(df, store, sales_reg, cust_reg)
        df.loc[df["Store"] == store, "SalesPred"] = pred_sales
        df.loc[df["Store"] == store, "CustPred"] = pred_cust
        stores_done += 1
        if (stores_done % print_progress_step == 0):
            print(f"{stores_done} stores done")
    print(f"{stores_done} stores done")
```

## Busqueda de hiper-parametros:

```python
from hyperopt import hp, tpe
from hyperopt.fmin import fmin
from hyperopt import Trials
import hyperopt
from lib import get_rmspe_score
```

```python
def objective(params):
    n_estimators = 10000
    clf = LGBMRegressor(
        n_estimators=n_estimators,
        n_jobs=-1,
        importance_type='gain',
        **params
    )
    clf.fit(X_train, np.ravel(y_train[0]), early_stopping_rounds=100, eval_set=[(X_val,np.ravel(y_val[0]))], eval_metric=light_gbm_rmspe_metric, verbose=0)
    y_pred = clf.predict(X_val)
    y_true = y_val[0].flatten()
    score = get_rmspe_score(y_true, y_pred)
    print("RMSPE:", score)
    return score
```

```python
import pickle
```

```python
space = {
    'max_depth': hp.choice('max_depth', list(np.arange(1, 15))),
    'num_leaves': hp.choice('num_leaves', list(np.arange(10, 50, 2))),
    'learning_rate': hp.loguniform('learning_rate', -6, -1),
    'reg_lambda': hp.loguniform('reg_lambda', -8, -3),
    'min_child_weight': hp.loguniform('min_child_weight', -8, -1),
    'min_child_samples': hp.choice('min_child_samples', list(np.arange(1, 30, 1))),
    'colsample_bytree': hp.choice('colsample_bytree', list(np.linspace(0.1, 1, 30))),
    'subsample': hp.choice('subsample', list(np.linspace(0.1, 1, 30)))
}

tpe_trials = Trials()

best = fmin(fn=objective,
            space=space,
            trials = tpe_trials,
            verbose=1,
            algo=hyperopt.tpe.suggest,
            return_argmin=False,
            max_evals=30)

pickle.dump(best, open('final/best_lgbm_params.pik', 'wb'))
pickle.dump(tpe_trials, open('final/best_lgbm_tpe.pik', 'wb'))
```
